```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.preprocessing.text as tfkpt
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
import json
```

```python
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
```

```python
headphones_df = pd.read_csv('ranked_headphones.csv')
```

```python
headphones = headphones_df.ReviewBody
sentiment = headphones_df.sentiment
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(headphones)
# tokenizer.get_config()
```

```python
headphones = tokenizer.texts_to_sequences(headphones)
headphones
```

```python
headphones = tokenizer.sequences_to_matrix(headphones)
headphones
```

```python
model = Sequential()

model.add(Dense(677, activation='sigmoid'))
model.add(Dropout(0.1))
model.add(Dense(557, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(281, activation='elu'))
model.add(Dropout(0.4))
model.add(Dense(167, activation='tanh'))
model.add(Dropout(0.5))
model.add(Dense(61, activation='softsign'))
model.add(Dense(3, activation='softmax'))
```

```python

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
sentiment = keras.utils.to_categorical(sentiment, 3)
```

```python
model.fit(headphones, sentiment, batch_size=37, epochs=7, verbose=1, validation_split=0.2, shuffle=True)
```
